```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import plotly
import plotly.express as px
```

```python
data = pd.read_csv('https://cdncontribute.geeksforgeeks.org/wp-content/uploads/nba.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
dataNBA =data
```

```python
dataNBA["Height"]=dataNBA["Height"].str.replace("-", ".")
dataNBA["Number"] =dataNBA["Number"].astype("object")
dataNBA["Height"] =dataNBA["Height"].astype("float64")
dataNBA.info()
```

```python
dataNBA.isnull().sum()
```

```python
dataNBA.head(50)
```

```python
dataNBA.tail(50)
```

```python
f, axes = plt.subplots(2, 2, figsize=(12, 12))

k=int(np.ceil(1+ np.log2(len(dataNBA))))
sns.distplot(dataNBA["Age"],kde=True, bins=k, ax=axes[0,0])
sns.distplot(dataNBA["Height"],kde=True, bins=k, color= "red", ax=axes[1,0])
sns.distplot(dataNBA["Weight"],kde=True, bins=k, color ="green", ax=axes[1,1])
sns.distplot(dataNBA["Salary"],kde=True, bins=k, color="purple", ax=axes[0,1])


plt.figure(figsize=(12,6))
sns.countplot(x="Team",data=dataNBA)
plt.xticks(rotation=90)

plt.figure(figsize=(12,6))
sns.countplot(x="Position", data=dataNBA);

plt.figure(figsize=(20,6))
sns.countplot(x="College", data=dataNBA);
plt.xticks(rotation=90)
```

```python
sns.heatmap(dataNBA.corr(), square=True, annot=True).set_title('Correlaciones')
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols #Ordinary linear squares

anova_NBA=ols('Salary ~ C(Number)+Team+College', data=dataNBA).fit()
sm.stats.anova_lm(anova_NBA)
```
