### 2.4 Análisis de valores atípicos

El comportamiento de las variables de este fenomeno no está muy ligado a los valores atípicos que se pueden observar en las variables, conociendo que esto no es el caso normal de cualquier fenomeno, para este caso en particular se comprobarán los ouliers por medio de la regla del rango intercuartil, de esta manera, si afectamos mucho el balance la variable obejtivo, se podrá concluír que los outliers son necesarios para modelar de manera correcta el modelo, de lo contrario, se procederá a remover los valores para mejorar el desempeño general del modelo.

En este caso, se espera que no haya una perdida de información importante respecto a la variable objetivo.




```python
f, axes = plt.subplots(2, 2, figsize=(9,9))
sns.boxplot(data['sibsp'], ax=axes[0,0])
sns.boxplot(data['age'], ax=axes[0,1])
sns.boxplot(data['parch'], ax=axes[1,0])
sns.boxplot(data['fare'], ax=axes[1,1])
plt.show()
```

#### Sobre los diagramas de caja:
* En el diagrama de caja se evidencian ouliers por regla del rango intercuartil en todas las variables
* El costo del ticket es la variable que más outliers presenta, como este valor esta directamente relacionado con la clase en la cual viajaban los pasajeros, y se presume que la misma tiene una alta correlación con la supervivencia de los ocupantes del barco, remover estos afectaría de manera significativa el desempeño del modelo.
* Para las otras variables, diferentes a "fare" se removerán los outliers presumiendo que no afectan de manera significativa el balance de la muestra


```python
## Definición de la función que removerá los outliers
def outliers_rem(data,column):
    '''Función creada para remover outliers definidos por la regla del rango intercuartil de un dataframe'''
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3-q1
    lower_bound = q1 -(1.5 * iqr) 
    upper_bound = q3 +(1.5 * iqr)
    data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return data
```

```python
## Removiendo outliers
dataFiltered = outliers_rem(data,'age')
dataFiltered = outliers_rem(dataFiltered,'parch')
dataFiltered = outliers_rem(dataFiltered,'sibsp')
```

```python
print(f'Cantidad de datos antes de filtrar outliers: {len(data)}')
print(f'Cantidad de datos después de filtrar outliers: {len(dataFiltered)}')
print(f'Se conserva el : {len(dataFiltered)*100/len(data)}% de los datos')
```

```python
f, axes = plt.subplots(2, 2, figsize=(9,9))
sns.boxplot(dataFiltered['sibsp'], ax=axes[0,0])
sns.boxplot(dataFiltered['age'], ax=axes[0,1])
sns.boxplot(dataFiltered['parch'], ax=axes[1,0])
sns.boxplot(dataFiltered['fare'], ax=axes[1,1])
plt.show()
```

Se evidencia que remover estos valores no afecta de manera drástica el balance en la variable objetivo para este fenómeno, por lo tanto, la remoción de los valores atípicos ayudará a mejorar el desempeño del modelo.
